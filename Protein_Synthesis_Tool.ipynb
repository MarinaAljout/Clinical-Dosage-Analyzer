```python
dna_seq = "ATGGCCATTGTAATGGGCCGCTGAAAGGGTCCCTAA"

mrna_seq = dna_seq.replace('T', 'U')
print("mRNA:", mrna_seq)
```

```python
genetic_code = {
    # Start Codon
    'AUG': {'name': 'Methionine', 'weight': 149.2},

    # Amino Acids
    'GCC': {'name': 'Alanine', 'weight': 89.1},
    'AUU': {'name': 'Isoleucine', 'weight': 131.2},
    'GUA': {'name': 'Valine', 'weight': 117.1},
    'GGC': {'name': 'Glycine', 'weight': 75.1},
    'CGC': {'name': 'Arginine', 'weight': 174.2},
    'AAA': {'name': 'Lysine', 'weight': 146.2},
    'CCG': {'name': 'Proline', 'weight': 115.1},
    'UCU': {'name': 'Serine', 'weight': 105.1},
    'ACC': {'name': 'Threonine', 'weight': 119.1},
    'UGG': {'name': 'Tryptophan', 'weight': 204.2},

    # Stop Codons
    'UAA': {'name': 'STOP', 'weight': 0},
    'UAG': {'name': 'STOP', 'weight': 0},
    'UGA': {'name': 'STOP', 'weight': 0}
}
```
