```python
def find_gene_segment(dna_sequence, start_motif, end_motif):

    dna_sequence = dna_sequence.upper()
    start_index = dna_sequence.find(start_motif)
    end_index = dna_sequence.find(end_motif)

    if start_index == -1 or end_index == -1:
        return "Warning: Pattern not found!"

    extracted_segment = dna_sequence[start_index : end_index + len(end_motif)]

    return extracted_segment
```

```python
def get_segment_info(segment):
    if "Warning" in segment:
        return segment

    info = {
        "Segment": segment,
        "Length": len(segment),
        "Is_Valid": "Yes" if len(segment) % 3 == 0 else "No (Potential Frameshift)" }
    return info
```

```python
dna_sequence = "GGCCATGCGATTACGATCGACTGAACCA"
start_motif = "ATGC"
end_motif = "TGA"

found_segment = find_gene_segment(dna_sequence, start_motif, end_motif)
final_report = get_segment_info(found_segment)

print(final_report)
```
